```python
import seaborn as sbn
import matplotlib.pyplot as plt
```

```python
sbn.get_dataset_names()
```

```python
mpg = sbn.load_dataset("mpg")
```

```python
mpg
```

### Introduction

```python
sbn.set_theme()
```

```python
sbn.set_style("darkgrid")
```

```python
sbn.set_context("paper")
```

```python
sbn.scatterplot(data=mpg, x="weight", y="mpg")
```

```python
a = mpg.corr(numeric_only=True)
a
```

```python
sbn.heatmap(a)
```

### Scatterplot

```python
plt.figure(figsize=(10, 7))
sbn.scatterplot(data=mpg, x="weight", y="mpg", hue="origin", palette="magma")
#sbn.despine(left=True)
```

### Line

```python
sbn.lineplot(data=mpg, x="model_year", y="weight")
```

### Point plot

```python
sbn.pointplot(data=mpg, x="model_year", y="weight")
```

### KDEPLOT

```python
sbn.kdeplot(data=mpg, x="model_year", y="mpg", fill=True)
```

```python
sbn.kdeplot(data=mpg, x="model_year")
```

### HistPlot


```python
sbn.histplot(data=mpg, x="weight", y="mpg")
```

```python
sbn.histplot(data=mpg, x="weight", kde=True)
```

### ECDPLOT

```python
sbn.ecdfplot(data=mpg, x="model_year")
```

### COUNTPLOT

```python
sbn.countplot(data=mpg, x="model_year")
```
